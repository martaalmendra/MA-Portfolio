### Segment Dataframe

We opted to develop a dataframe that included all segments to make it easier to implement association rules later.

```python
# Group the dataframe by clusters
persons_clust_grouped = persons_clust.groupby('Final_Cluster')

# Create a dataframe for each cluster
cluster_dataframes = {cluster: group for cluster, group in persons_clust_grouped}

# Remove the cluster 'Pet Lovers' since it has no information in customer_basket
del cluster_dataframes['Pet lovers']

promo_hunters = cluster_dataframes.get('Promo Hunters')
high_spenders = cluster_dataframes.get('High spenders')
vegetarians = cluster_dataframes.get('Vegetarians')
tech_enthusiasts = cluster_dataframes.get('Tech enthusiasts')
demanding_foodies = cluster_dataframes.get('Demanding foodies')
large_families = cluster_dataframes.get('Large families')
college_students = cluster_dataframes.get('College students')
fishy = cluster_dataframes.get('Fishy')
```
